```python
from google import genai
client = genai.Client(api_key=os.environ['GEMINI_API_KEY'])

resp = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='Explain the 5-Layer AI Skill Pyramid in 5 short bullet points.'
)
print(resp.text)
```

The 5-Layer AI Skill Pyramid outlines the foundational to strategic skills needed for successful AI implementation, from bottom to top:

*   **Data & MLOps Foundation:** The bedrock of infrastructure, data pipelines, and MLOps tools to ensure data readiness and reliable model deployment.
*   **Machine Learning Engineering:** Developing, training, and deploying ML models into production systems efficiently and scalably.
*   **Machine Learning Science & Research:** Innovating and experimenting with new algorithms, models, and techniques to advance the field.
*   **Domain Expertise & Business Acumen:** Applying AI solutions to specific industry problems, understanding their real-world impact and value.
*   **AI Strategy & Leadership:** Defining the vision, ethical guidelines, and strategic roadmap for AI initiatives within an organization.
